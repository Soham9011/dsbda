```python
def fun1(value):
    if (value == "male"):
        return 1
    else:
        return 0
```

```python
def fun2(value):
    if (value == 'S'):
        return 0
    elif (value == 'C'):
        return 1
    elif (value == 'Q'):
        return 2
    else:
        return 0
```

```python
data["Sex"] = data["Sex"].apply(fun1)
```

```python
data["Embarked"] = data["Embarked"].apply(fun2)
```

```python
data.isnull().sum()
```

```python
data = data.drop("Cabin", axis=1)
```
